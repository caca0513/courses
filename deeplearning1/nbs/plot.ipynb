```python
%matplotlib notebook

from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
import math as math
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(0,10)
y = np.arange(-10,-5)
xx, yy = np.meshgrid(x, y)
z = [0]*xx
```

```python
x, y, xx, yy, z
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(xx, yy, z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```
